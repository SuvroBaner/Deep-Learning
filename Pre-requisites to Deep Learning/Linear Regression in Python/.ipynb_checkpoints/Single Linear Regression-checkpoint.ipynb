# 1D solution for Linear Regression. Estimating a and b

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the data
X = [] # a list for X vector (independent variable)
Y = [] # a list for Y vector (dependent variable)

for line in open('data_1d.csv'):
    x, y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
```

```python
# Let's turn them into NumPy array
X = np.array(X)
Y = np.array(Y)
```

```python
# Plot to see what it looks like-
plt.scatter(X, Y)
plt.xlabel('Value of X')
plt.ylabel('Value of Y')
plt.title('Plotting the data')
plt.show()
```

```python
# Apply the equations we learned to calculate a and b which we get from minimizing the error function (cost or objective fn)
```

```python
# We have the same denominator for both the equations of a and b
# The first term is summation xi squared can be also written as dot product of X vector with itself.
# The second term is mean of X times the summation of X, this Xs are vectors 
denominator = X.dot(X) - X.mean()*X.sum()

# First term is the summation of xi and yi which is same as dot product of X and Y where both are vectors.
a = (X.dot(Y) - Y.mean()*X.sum()) / denominator

b = (Y.mean()*X.dot(X) - X.mean()*Y.dot(X)) / denominator
```

```python
# Calculate predicted Y
Yhat = a*X + b
```

```python
# Plot it all
plt.scatter(X, Y)
plt.plot(X, Yhat)
plt.xlabel('Value of X')
plt.ylabel('Value of Y')
plt.title('Plotting after the linear regression')
plt.show() # The plot looks like good fit to the points
```

```python
# Now let's understand how did my model do. We will compute the R-squared statistics for that.
d1 = Y - Yhat  # both are vectors and the difference is still a vector
d2 = Y - Y.mean() # here Y is a vector and the mean of Y is a scalar, when we subtract the mean of Y is subtracted from all the 
# elements of the Y vector

r2 = 1 - d1.dot(d1) / d2.dot(d2)
print("The R-squared is :", r2)
```

```python
# The r-squared is pretty close to 1 which means that the data is almost a perfect fit.
# It is so because we have cooked up the data for this example
```

# Demonstrating Moore's Law using Simple Linear Regression

```python
# Moore's Law : Transistor Count on Integrated Circuits doubles every two years
# T0 = A0 ; T1 = 2A0 ; T2 = 2(2A0) ... ; Tn = 2^nA0
```

```python
import re
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the data
X = []
Y = []

non_decimal = re.compile(r'[^\d]+') # matches a digit

for line in open('moore.csv'):
    r = line.split('\t')
    x = int(non_decimal.sub('', r[2].split('[')[0]))  # we get the year. we are also substituting '[' by a blank (wikipedia ref)
    y = int(non_decimal.sub('', r[1].split('[')[0])) # we get the transistor count
    X.append(x)
    Y.append(y)
    
# Converting them to a NumPy array-
X = np.array(X)
Y = np.array(Y)

# Now we will do the log transformation of the Moore's low equation which he has hypothesized
# log(Tn) = nlog(log 2) + log(A0)

logY = np.log(Y) # log Y makes it linear as we can see above

# Plotting them-
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].scatter(X, Y)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title("Plot Transistor Count vs Time ")

axes[1].scatter(X, logY)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Log Y')
axes[1].set_title("Plot Log of Transistor Count vs Time")
```

```python
# Now estimating a and b using linear regression-

Y = np.log(Y)  # taking the log to make Y vector linear 

denominator = X.dot(X) - X.mean()*X.sum()
a = (X.dot(Y) - Y.mean()*X.sum()) / denominator
b = (Y.mean()*X.dot(X) - X.mean()*Y.dot(X)) / denominator

# Calculate predicted Y
Yhat = a*X + b

# Plot it all
plt.scatter(X, Y)
plt.plot(X, Yhat)
plt.xlabel('Value of X')
plt.ylabel('Value of Log Y')
plt.title('Plotting after the linear regression')
plt.show() # The plot looks like good fit to the points
```

```python
# Now let's understand how did my model do. We will compute the R-squared statistics for that.
d1 = Y - Yhat  
d2 = Y - Y.mean() 
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print("The R-squared is :", r2)
```

```python
# Now, figure out how long does it take for the transisitor count to get doubled.
# So, our linear regression was : log(tc) = a*year + b  where tc = transistor count
# now, when we untransoform it : tc = exp(a*year) * exp(b)
# Then, 2*tc = 2*exp(a*year) * exp(b) = exp(ln(2))*exp(a*year) * exp(b)
#            = exp(b) * exp(a*year + ln(2))

# So, L.H.S : exp(b) * exp(a*year2)
# and R.H.S : exp(b) * exp(a*year1 + ln(2))
# and L.H.S = R.H.S
# So, a*year2 = a*year1 + ln(2)  # just simplifying
# => year2 = year1 + ln(2)/a

# So, time to double-
print('Time to double: ', np.log(2)/a, " years")
```
